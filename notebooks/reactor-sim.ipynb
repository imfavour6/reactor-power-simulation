## Project Overview

This project investigates reactor power behavior using computational modeling in Python. The simulation explores the effects of reactivity changes, control rod movement, and temperature feedback on reactor power over time.

The goal is to combine nuclear engineering principles with data science techniques for analysis and visualization.

## Reactivity

```python
#Example for understanding 

reactivity  = - 0.2

if reactivity > 0:
    print("Positive reactivity, Power Increasing ")
elif reactivity < 0:
    print("Negative reactivity, Power Decreasing ")
else:
    print("Zero reactivity, Power Constant ")
```

```python
#starting neutrons in the reactor
neutrons =  1000

# reactivity
reactivity = 0.2

# Calculate the change in neutrons based on reactivity
change_in_neutrons = neutrons * reactivity

# Calculate the new number of neutrons after the change
new_neutrons = neutrons + change_in_neutrons

print("Initial neutrons:", neutrons)
print ("Change in neutrons:", change_in_neutrons) 
print("New neutrons:", new_neutrons)
```

```python
list = []
for time in range(1,10):
    change_in_neutrons = neutrons * reactivity # Calculate the change in neutrons based on reactivity
    new_neutrons = neutrons + change_in_neutrons # Calculate the new number of neutrons after the change
    neutrons = new_neutrons  # Update neutrons for the next iteration
    print('Time:', time, 'Neutrons:', neutrons)
    list.append(neutrons)

```
